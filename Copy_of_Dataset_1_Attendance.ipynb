```python
attendance = """
  SELECT
    report_period,
    school_code,
    student_class,
    subject_code,
    electronic_diary,
    FIRST(student_count_diary) AS student_count,
    SUM(excused_lessons_illness) AS excused_lessons_illness,
    SUM(excused_lessons_other) AS excused_lessons_other,
    SUM(unexcused_lessons) AS unexcused_lessons
  FROM 'attendance/attendance.parquet'
  WHERE division_code IS NULL
  GROUP BY all
"""
```

Final query

```python
query = f"""

WITH
  attendance AS ({attendance}),
  school AS ({school}),
  subject AS ({subject})

SELECT
  school.school_name,
  school.municipality_name,
  subject.subject_name_en,
  attendance.student_class,
  attendance.report_period,
  attendance.student_count,
  attendance.excused_lessons_illness,
  attendance.excused_lessons_other,
  attendance.unexcused_lessons
FROM attendance
JOIN school ON attendance.school_code = school.school_code -- == USING (school_code)
JOIN subject on attendance.subject_code = subject.subject_code
  AND attendance.electronic_diary = subject.electronic_diary
"""
print(query)
```

```python
duckdb.sql(query).df()
```
